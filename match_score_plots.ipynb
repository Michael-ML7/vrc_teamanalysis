```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('1381P_matches.csv')
```

```python
def plot_all_matches():
    x_vals = df['Opponent Score']
    y_vals = df['Team Score']
    colors = ['blue' if x > y else 'red' for x, y in zip(x_vals, y_vals)]

    plt.figure(figsize=(8, 8))
    plt.scatter(x_vals, y_vals, c=colors)
    plt.axhline(0, color='gray', linewidth=0.8)
    plt.axvline(0, color='gray', linewidth=0.8)
    plt.xlabel('Opponent Score (Blue)')
    plt.ylabel('Team Score (Red)')
    plt.title('All Matches: Team Score vs Opponent Score')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xticks(range(0, int(max(x_vals.max(), y_vals.max())) + 5, 5))
    plt.yticks(range(0, int(max(x_vals.max(), y_vals.max())) + 5, 5))
    plt.show()
    
plot_all_matches()
```

```python
def plot_filtered_event_types():
    event_filter = ['World', 'National', 'Regional', 'State', 'Signature']
    filtered_df = df[df['Event Type'].isin(event_filter)]
    x_vals = filtered_df['Opponent Score']
    y_vals = filtered_df['Team Score']
    colors = ['blue' if x > y else 'red' for x, y in zip(x_vals, y_vals)]

    plt.figure(figsize=(8, 8))
    plt.scatter(x_vals, y_vals, c=colors)

    # Label each point with its coordinates
    for x, y in zip(x_vals, y_vals):
        plt.text(x, y, f"({x},{y})", fontsize=8, ha='right', va='bottom')

    plt.axhline(0, color='gray', linewidth=0.8)
    plt.axvline(0, color='gray', linewidth=0.8)
    plt.xlabel('Opponent Score (Blue)')
    plt.ylabel('Team Score (Red)')
    plt.title('Filtered Event Types: Team Score vs Opponent Score')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xticks(range(0, int(max(x_vals.max(), y_vals.max())) + 5, 5))
    plt.yticks(range(0, int(max(x_vals.max(), y_vals.max())) + 5, 5))
    plt.show()

plot_filtered_event_types()
```

```python
def plot_filtered_event_types_no_qualifier():
    event_filter = ['World', 'National', 'Regional', 'State', 'Signature']
    filtered_df = df[df['Event Type'].isin(event_filter)]
    filtered_df = filtered_df[~filtered_df['Match Name'].str.contains('qualifier', case=False, na=False)]
    x_vals = filtered_df['Opponent Score']
    y_vals = filtered_df['Team Score']
    colors = ['blue' if x > y else 'red' for x, y in zip(x_vals, y_vals)]

    plt.figure(figsize=(8, 8))
    plt.scatter(x_vals, y_vals, c=colors)

    # Label each point with its coordinates
    for x, y in zip(x_vals, y_vals):
        plt.text(x, y, f"({x},{y})", fontsize=8, ha='right', va='bottom')

    plt.axhline(0, color='gray', linewidth=0.8)
    plt.axvline(0, color='gray', linewidth=0.8)
    plt.xlabel('Opponent Score (Blue)')
    plt.ylabel('Team Score (Red)')
    plt.title('Filtered Events w/o "Qualifier": Team Score vs Opponent Score')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xticks(range(0, int(max(x_vals.max(), y_vals.max())) + 5, 5))
    plt.yticks(range(0, int(max(x_vals.max(), y_vals.max())) + 5, 5))
    plt.show()

plot_filtered_event_types_no_qualifier()
```
